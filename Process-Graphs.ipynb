# Common Processes in Stochastic Analysis

```python
import itopy
import matplotlib.pyplot as plt
import numpy as np
```

### Wiener Process (Classical Brownian Motion)

The stochastic differential equation (SDE) for Brownian motion (optionally with drift and volatility) can be expressed as:

**dX_t = γ dt + σ dW_t**

Where:
- **X_t** is the stochastic process.
- **γ** is the drift coefficient.
- **σ** is the volatility coefficient.
- **W_t** is the Wiener process (standard Brownian motion).

```python
x0 = 0
n_steps = 1000
T = 1
gamma = 0
sigma = 1

times1, brown1 = itopy.wiener(x0,n_steps,T,gamma,sigma);
times2, brown2 = itopy.wiener(0,1000,1,5,1);
times3, brown3 = itopy.wiener(0,1000,1,0,2);

plt.plot(times1,brown1,label='Classic Brownian motion');
plt.plot(times2,brown2,label='Brownian motion with postive drift');
plt.plot(times3,brown3,label='Brownian motion with volatility of 2');
plt.xlabel('Time')
plt.ylabel('W_t')
plt.legend()
plt.show()
```

### Geometric Brownian Motion (GBM)
The stochastic differential equation (SDE) for Geometric Brownian Motion can be expressed as:
**dS_t = μ S_t dt + σ S_t dW_t**

Where:
- **S_t** is the stochastic process (e.g., stock price).
- **μ** is the drift coefficient (rate of return).
- **σ** is the volatility coefficient.
- **W_t** is the Wiener process (standard Brownian motion).

```python
x0 = 100
n_steps = 100
T = 1
mu = 0.2
sigma = 1

times1, brown1 = itopy.geo_brownian(x0,n_steps,T,mu,sigma);
times2, brown2 = itopy.geo_brownian(x0,n_steps,T,mu,sigma);
times3, brown3 = itopy.geo_brownian(x0,n_steps,T,mu,sigma);



plt.plot(times1,brown1,label='Random Path 1');
plt.plot(times2,brown2,label='Random Path 2');
plt.plot(times3,brown3,label='Random Path 3');
plt.xlabel('Time')
plt.ylabel('S_t')
plt.legend()
plt.show()
```

### Brownian Bridge

Brownian motion conditioned to start and end at zero.

**B(t) = W(t) - (t/T) * W(T)**

Where:
- **W(t)**: Standard Brownian motion at time **t**.
- **T**: Total time interval.
- **W(T)**: Value of Brownian motion at time **T**.


```python
x0 = 100
x1 = 100
n_steps = 500
T = 1

times1, bridge1 = itopy.brownian_bridge(x0,x1,n_steps,T);
times2, bridge2 = itopy.brownian_bridge(x0,x1,n_steps,T);
times3, bridge3 = itopy.brownian_bridge(x0,x1,n_steps,T);


plt.plot(times1,bridge1,label='Sample 1');
plt.plot(times2,bridge2,label='Sample 2');
plt.plot(times3,bridge3,label='Sample 3');
plt.xlabel('Time')
plt.ylabel('B_t')
plt.legend()
plt.show()
```

### Maximum Process of Brownian Motion

The maximum process of Brownian motion tracks the maximum value attained by a Brownian motion up to time **t**. It is defined as:

**M_t = max{W_s : 0 ≤ s ≤ t}**

Where:
- **M_t** is the maximum process of Brownian motion.
- **W_s** is the Wiener process (standard Brownian motion) at time **s**.


```python
x0 = 0
n_steps = 500
T = 1
gamma = 0
sigma = 1

times1, maxW = itopy.max_process(x0,n_steps,T,gamma,sigma);

plt.plot(times1,maxW,label='Maximum of Brownian Motion');

plt.xlabel('Time')
plt.ylabel('M_t')
plt.legend()
plt.show()
```
